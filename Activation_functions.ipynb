# **Sigmoid**

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
  return 1/(1+np.exp(-x))

```

```python
x = np.linspace(-20,20,100)
print(x)
plt.plot(x,sigmoid(x))
plt.axis('tight')
plt.title('Activation Function :Sigmoid')
plt.show()
```

# **tanh**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def tanh(x):
  return np.tanh(x)
x = np.linspace(-10, 10,100)
plt.plot(x, tanh(x))
plt.axis('tight')
plt.title('Activation Function :tanh')
plt.show()
```

# **ReLU**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def RELU(x):
  x1=[]
  for i in x:
    if i<0:
      x1.append(0)
    else:
      x1.append(i)
  return x1
```

```python
x = np.linspace(-10, 10,100)
plt.plot(x, RELU(x))
plt.axis('tight')
plt.title('Activation Function :ReLU')
plt.show()
```

# **Softmax**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def softmax(x):
  return np.exp(x) / np.sum(np.exp(x), axis=0)
```

```python
x = np.linspace(-10, 10)
plt.plot(x, softmax(x))
plt.axis('tight')
plt.title('Activation Function :Softmax')
plt.show()
```
