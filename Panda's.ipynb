```python
# Merging
t1 = {
    'subject_id': ['1', '2', '3', '4', '5'],
    'student_name': ['Gopi', 'F1', 'F2', 'F3', 'F4']
}

t2 = {
    'subject_id': ['4', '5', '6', '7', '8'],
    'student_name': ['Varun', 'F5', 'F6', 'F7', 'F8']
}

df1 = pd.DataFrame(t1, columns=['subject_id', 'student_name'])
df2 = pd.DataFrame(t2, columns=['subject_id', 'student_name'])
print(df1)
print(df2)
```

```python
pd.merge(df1,df2,on='subject_id')
```
